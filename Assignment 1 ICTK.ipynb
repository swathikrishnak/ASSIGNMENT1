### 2. scatter plot 

```python

sns.scatterplot(data['SL'],data['SW'],hue=data['Classification'])#s=size of mark
plt.xlabel('SL')
plt.ylabel('SW')
plt.show()
```

OBSERVATION
1. Using sepal length and sepal width features, we can distinguish setosa from others.
2. Seperating vesicolor from viginica is much harder as they considerable overlap.

```python
sns.scatterplot(data['PL'],data['PW'],hue=data['Classification'])#s=size of mark
plt.xlabel('PL')
plt.ylabel('PW')
plt.show()
```

OBSERVATION
1. Using petal length and  petal width features, one class is entirely isolated from other.
2. There is only few overlap between  virginica and versicolor so they can be seperated easily.
